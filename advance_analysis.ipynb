```python
# Total Social Spending
df["Total_Social_Spending"] = (
    df["Education_Amount_Billions_USD"] +
    df["Health_Amount_Billions_USD"] +
    df["Social_Welfare_Amount_Billions_USD"]
)

# Development Spending
df["Development_Spending"] = (
    df["Infrastructure_Amount_Billions_USD"] +
    df["Agriculture_Amount_Billions_USD"]
)

# Governance Spending
df["Governance_Spending"] = (
    df["Defense_Amount_Billions_USD"] +
    df["Administration_and_Others_Amount_Billions_USD"] +
    df["Interest_Payments_Amount_Billions_USD"]
)
```

```python
# Budget Utilization (%)
df["Social_Spending_%"] = (
    df["Total_Social_Spending"] /
    df["Total_Budget_Billions_USD"]
) * 100

df["Development_Spending_%"] = (
    df["Development_Spending"] /
    df["Total_Budget_Billions_USD"]
) * 100

df["Governance_Spending_%"] = (
    df["Governance_Spending"] /
    df["Total_Budget_Billions_USD"]
) * 100
```

```python
# Total Allocated Amount
df["Total_Allocated"] = (
    df["Defense_Amount_Billions_USD"] +
    df["Education_Amount_Billions_USD"] +
    df["Health_Amount_Billions_USD"] +
    df["Interest_Payments_Amount_Billions_USD"] +
    df["Infrastructure_Amount_Billions_USD"] +
    df["Agriculture_Amount_Billions_USD"] +
    df["State_Transfers_Amount_Billions_USD"] +
    df["Social_Welfare_Amount_Billions_USD"] +
    df["Administration_and_Others_Amount_Billions_USD"]
)

# Budget Remaining
df["Budget_Remaining"] = (
    df["Total_Budget_Billions_USD"] -
    df["Total_Allocated"]
)
```
