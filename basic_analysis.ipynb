```python
import matplotlib.pyplot as plt

sectors = [
    "Defense",
    "Education",
    "Health",
    "Infrastructure",
    "Agriculture",
    "State Transfers",
    "Social Welfare",
    "Administration"
]

amounts = [
    df["Defense_Amount_Billions_USD"].sum(),
    df["Education_Amount_Billions_USD"].sum(),
    df["Health_Amount_Billions_USD"].sum(),
    df["Infrastructure_Amount_Billions_USD"].sum(),
    df["Agriculture_Amount_Billions_USD"].sum(),
    df["State_Transfers_Amount_Billions_USD"].sum(),
    df["Social_Welfare_Amount_Billions_USD"].sum(),
    df["Administration_and_Others_Amount_Billions_USD"].sum()
]

plt.figure(figsize=(10,5))
plt.bar(sectors, amounts)
plt.xticks(rotation=45)
plt.ylabel("Budget (Billion USD)")
plt.title("Sector-wise Budget Allocation")
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(
    amounts,
    labels=sectors,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Budget Distribution by Sector")
plt.show()
```

```python
import matplotlib.pyplot as plt

corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))
plt.imshow(corr, aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```
